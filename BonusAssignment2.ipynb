```python
#Import data using Pandas
import pandas as pd

input = '/content/tested.csv'
df = pd.read_csv(input)
#First 5 rows
print(df.head())
import seaborn as sns
import matplotlib.pyplot as plt

#Handling Missing Values
print(df.isnull().sum())
```

```python
#EDA - Exploratory Data Analysis
#Handling Missing values

print("Data Types of Column")
print(df.dtypes)
#Identifying the data type & format of dataset
df = df.dropna()
print(df.isnull().sum())
print("\n")

```

```python
# Select only numeric columns for correlation matrix
numeric_columns = df.select_dtypes(include=['number']).columns
df_numeric = df[numeric_columns]
# Compute the correlation matrix
correlation_matrix = df_numeric.corr()
# Display the correlation matrix
print("\nCorrelation Matrix:")
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Count plot for 'Survived' column, updated to avoid FutureWarning
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Survived', palette='Set2', hue=None) # Avoiding FutureWa
plt.title('Survival Count in Titanic Dataset')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
print("Visualizing the Survival data in Titanic")

plt.show()
```
